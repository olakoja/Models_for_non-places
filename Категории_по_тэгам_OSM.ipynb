```python
# Определяем порядок обработки ключей
priority_keys = ('name', 'amenity', 'building', 'landuse', 'historic', 'leisure', 'man_made', 'military', 'natural', 'office', 'place', 'shop', 'sport', 'tourism', )

def find_non_place_value(tags, row):
    """Находит первое соответствие в словаре tags для строки row."""
    # Проверяем ключ name
    if 'name' in tags and 'name' in row and row['name'] is not None:
        name_string = row['name']
        # Проверяем, содержатся ли в name определенные слова
        invalid_values = ['улица','проспект','мост' , 'Магнит', 'Пятёрочка', 'Пятерочка', 'Аэродром', 'Детский сад', 'Школа', 'завод',
                          'корпус', 'Центр', 'магазин',  'дом', 'сад', 'парк', 'сквер', 'церковь', 'собор', 'Церковь', 'Собор', 'храм', 'Храм ', ' ЖК ']
        if any(value in name_string for value in invalid_values):
            return True  # Если содержится, возвращаем True

    # Проверяем приоритетные ключи
    for priority_key in priority_keys[1:]:  # Пропускаем 'name'
        if priority_key in tags and priority_key in row and row[priority_key] is not None:
            tag_list = tags[priority_key]
            if tag_list is not None:
                if isinstance(tag_list, list) and row[priority_key] in tag_list:
                    return True
                elif row[priority_key] == tag_list:
                    return True
    return False

def check_other_keys(tags, row):
    """Проверяет все остальные ключи в словаре tags."""
    for tag_type, tag_list in tags.items():
        if tag_type in priority_keys:
            continue
        if tag_list is not None and tag_type in row and row[tag_type] is not None:
            if isinstance(tag_list, list) and row[tag_type] in tag_list:
                return True
            elif row[tag_type] == tag_list:
                return True
    return False

# Проверка каждого объекта
for index, row in geodata.iterrows():
    non_place_value = None  # Переменная для хранения первого найденного соответствия

    for dict_name in dictionary_names:
        tags = globals()[dict_name]

        # Проверяем ключ name
        if find_non_place_value(tags, row):
            non_place_value = dict_name
            break

        # Проверяем все остальные ключи
        if check_other_keys(tags, row):
            non_place_value = dict_name
            break

    # Записываем первое найденное значение в "non-place"
    if non_place_value:
        geodata.at[index, 'non-place'] = non_place_value
```

```python
# Уникальные значеничя
unique_values = geodata['non-place'].unique()

print("Уникальные значения в столбце 'non-place':")
print(unique_values)

unique_values = geodata['non-place'].value_counts()
уникальные_значения_df = geodata['non-place'].value_counts().reset_index()
уникальные_значения_df.columns = ['значение', 'количество']
print(уникальные_значения_df)
```

```python
# Сохранение обновленных геоданных
geodata.to_file('updated_geodata.gpkg', driver="GPKG", index=False)
```

```python
# Замените 'your_geodata_file.csv' на путь к вашему файлу
geodata = gpd.read_file('/content/границы.gpkg')

# Создание GeoDataFrame
gdf = gpd.GeoDataFrame(geodata)

# Функция для извлечения ключей и значений из 'other_tags'
def extract_keys_values(row):
    if pd.isna(row['other_tags']):
        return []

    tags = row['other_tags'].split(',')
    key_value_pairs = []

    for tag in tags:
        key_value = tag.split('=>')
        key = key_value[0].strip().strip('"')  # Убираем кавычки и пробелы
        value = key_value[1].strip().strip('"') if len(key_value) > 1 else None
        key_value_pairs.append((key, value))

    return key_value_pairs

# Словарь для хранения уникальных ключей
unique_keys = set()

# Извлечение ключей и значений
for index, row in gdf.iterrows():
    key_value_pairs = extract_keys_values(row)
    for key, value in key_value_pairs:
        unique_keys.add(key)

# Создание новых столбцов для каждого уникального ключа
for key in unique_keys:
    gdf[key] = None  # Инициализируем новый столбец значениями None

    # Заполнение значений в новом столбце
    for index, row in gdf.iterrows():
        key_value_pairs = extract_keys_values(row)
        for k, value in key_value_pairs:
            if k == key:
                gdf.at[index, key] = value  # Заполняем значение в новом столбце
                break  # Прерываем цикл, если нашли соответствие
```
